```python
import os
import pandas as pd
```

```python
file = 'Tomofun-AI--master/train'
```

```python
df = pd.read_csv(os.path.join(file, 'meta_train.csv'))
df
```

```python
df['Remark'].unique()
```

```python
classes_names = ['Barking', 'Howling', 'Crying', 'COSmoke', 'GlassBreaking', 'Other']
```

```python
def create_label_dict():
    label_dict = {}
    for i, label_name in enumerate(classes_names):
        label_dict[label_name] = i

    return label_dict 
```

```python
label_dict = create_label_dict()

label_dict
```

```python
data = {}

for i in range(len(df['Filename'])):
    fname = df['Filename'][i] + '.wav'
    label = df['Remark'][i]
    
    if df['Remark'][i] not in label_dict:
        label = 'Other'
    
    data[fname] = label
```

```python
data
```

```python
new_df = pd.DataFrame(list(data.items()), columns=["fname", "label"])
new_df
```

```python
new_df.to_csv("tomofun_train.csv", index=False)
```
